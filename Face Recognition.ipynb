```python
import face_recognition
from skimage.feature import hog
from skimage import data, exposure
import cv2

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
import numpy as np

import os
%matplotlib inline
```

```python
path = r'C:\Users\Desktop\Diacto1\Modi'
```

```python
image = cv2.imread(os.path.join(path, 'trumpkim.jpg'))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
fd, hog_image = hog(image, orientations = 8, pixels_per_cell = (16, 16), cells_per_block = (1, 1), visualize = True, multichannel = True)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (9,5), sharex = True, sharey = True)
ax1.axis('off')
ax1.imshow(image, cmap = plt.cm.gray)
ax1.set_title('Input Image')

hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range = (0, 10))

ax2.axis('off')
ax2.imshow(hog_image_rescaled, cmap = plt.cm.gray)
ax2.set_title('Histogram Of Oriented Gradients[HOG] image')
plt.show()
```

```python
print("Original :",  image.shape)
print("HOG :", len(fd))
```

```python
face_locations = face_recognition.face_locations(image)
number_of_faces = len(face_locations)
```

```python
print("Found {} face(s) in the input image".format(number_of_faces))
```

```python
plt.imshow(image)
ax = plt.gca()

for i in face_locations:
    
    top, right, bottom, left = i
    l, r, t, b = left, right, top, bottom
    print(f"Following face has been detected at pixel position : Left: {l}, Right: {r}, Top: {t}, Bottom: {b}")
    
    rect = Rectangle((l, t), r - l, b - t, fill = False, color = 'yellow')
    ax.add_patch(rect)
    
plt.show()
```

# Small Practical Application
## With more Images and their Processing

```python
image1 = cv2.imread(os.path.join(path, 'N_MODI.jpg'))
N_MODI = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

image2 = cv2.imread(os.path.join(path, 'AMIT_SHAH.jpg'))
A_SHAH = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

image3 = cv2.imread(os.path.join(path, 'SAMBIT_PATRA.jpg'))
S_PATRA = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (9,5), sharex = True, sharey = True)
ax1.axis('off')
ax1.imshow(image1, cmap = plt.cm.gray)
ax1.set_title('Original Image')

ax2.axis('off')
ax2.imshow(N_MODI, cmap = plt.cm.gray)
ax2.set_title('Converted Image')
plt.show()
```

## Creating the encodings of multiple images.

```python
N_MODI_encoding = face_recognition.face_encodings(N_MODI)[0] # [0] : In order to contain multiple faces.

A_SHAH_encoding = face_recognition.face_encodings(A_SHAH)[0]

S_PATRA_encoding = face_recognition.face_encodings(S_PATRA)[0]
```

## Saving the encodings in a list to create 'Known_Encodings' database.

```python
Known_Encodings = [N_MODI_encoding, A_SHAH_encoding, S_PATRA_encoding] # To ready our database.
```

# Unknown Image + Testing
## Narendra Modi Face Recognition

```python
Unknown_Image = cv2.imread(os.path.join(path, 'Laughing Modi.jpg'))   # Tried Parameters : 'Laughing Modi.jpg', Sambit Patra Testing.jpg, 'Amit Shah Testing.jpg'
Unknown_Image = cv2.cvtColor(Unknown_Image, cv2.COLOR_BGR2RGB)
plt.imshow(Unknown_Image)
```

```python
Unknown_face_encodings = face_recognition.face_encodings(Unknown_Image) # Not using [0] at end because unknown image can contain multiple faces to be recognized.
```

```python
from scipy.spatial import distance   # For computing Euclidean Distance.

for i in Unknown_face_encodings:
    
    results = []
    
    for j in Known_Encodings:
        d = distance.euclidean(j, i)
        
        results.append(d)
        
    threshold = 0.6
    results = np.array(results) <= threshold
    
    name = "Unknown"
    
    if results[0]:
        name = "Narendra Modi Face"
        
    elif results[1]:
        name = "Amit Shah Face"
        
    elif results[2]:
        name = "Sambit Patra Face"
        
    print(f'Found {name} in the image!')
        
```
